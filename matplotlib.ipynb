```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['PSSM+PscAAC', 'PscAAC+TPC', 'PSSM+TPC', 'PSSM+PscAAC+TPC']
metrics = ['AUC', 'Acc (%)', 'Sen (%)', 'Spe (%)', 'MCC', 'Precision', 'F1 measure']

# Example data (replace with actual values)
data = np.array([
    [0.95, 90, 85, 88, 0.87, 0.89, 0.88],  # PSSM+PscAAC
    [0.93, 88, 84, 86, 0.85, 0.87, 0.86],  # PscAAC+TPC
    [0.94, 89, 86, 87, 0.86, 0.88, 0.87],  # PSSM+TPC
    [0.96, 91, 87, 89, 0.88, 0.90, 0.89]   # PSSM+PscAAC+TPC
])

x = np.arange(len(metrics))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))

# Plotting the bars for each method
for i, method in enumerate(methods):
    ax.bar(x + i*width, data[i], width, label=method)

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Comparison of Different Methods')
ax.set_xticks(x + width * (len(methods) - 1) / 2)
ax.set_xticklabels(metrics)
ax.legend()

fig.tight_layout()

plt.show()
```
