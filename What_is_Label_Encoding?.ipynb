```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"name":["wscube","cow","cat","Dog","black"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["en_name"] = le.fit_transform(df["name"])
```

```python
df
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset['Property_Area'].unique()
```

```python
la = LabelEncoder()
la.fit_transform(dataset["Property_Area"])
```

```python
dataset["en_Property_Area"] = la.fit_transform(dataset["Property_Area"])
```

```python
dataset
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(dataset['Gender'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(dataset, x='ApplicantIncome', y='Gender', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```
