```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
df = pd.concat([train, test],sort=False)
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Age', y='SibSp', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(['Cabin','Survived'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python

# Remove outliers from 'Fare'
df['Fare_zscore'] = zscore(df['Fare'])
threshold = 3
df = df[df['Fare_zscore'] <= threshold]
df.drop(columns=['Fare_zscore'], inplace=True)

display(df.head())
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Encode categorical features
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])

# Split data back into train and test
train_df = df[df['PassengerId'].isin(train['PassengerId'])]
test_df = df[df['PassengerId'].isin(test['PassengerId'])]

# Define features and target
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X_train = train_df[features]
y_train = train[train['PassengerId'].isin(train_df['PassengerId'])]['Survived']
X_test = test_df[features]

# Initialize and train the AdaBoost classifier
adaboost = AdaBoostClassifier(n_estimators=100, random_state=42)
adaboost.fit(X_train, y_train)

# Make predictions
y_pred = adaboost.predict(X_test)

# Create submission file
submission = pd.DataFrame({
    "PassengerId": test_df["PassengerId"],
    "Survived": y_pred
})
submission.to_csv('submission.csv', index=False)

display(submission.head())
```
