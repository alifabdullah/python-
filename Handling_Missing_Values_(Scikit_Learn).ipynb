```python
si = SimpleImputer(strategy="mean")
ar = si.fit_transform(df[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

```python
new_df = pd.DataFrame(ar,columns = df.select_dtypes(include = "float64").columns)
```
